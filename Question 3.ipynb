## Method 1 - Boruta Feature Selection

```python
#libraries
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LassoCV, LinearRegression


import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler

import warnings
warnings.filterwarnings('ignore')

pd.set_option("display.max_columns", 500)
```

```python
# Function for ranking variables
def ranking(ranks, names, order=1):
    minmax = MinMaxScaler()
    ranks = minmax.fit_transform(order * np.array([ranks]).T).T[0]
    ranks = map(lambda x: round(x, 2), ranks)
    return dict(zip(names, ranks))

# Read dataset
df = pd.read_csv("cases_state.csv")

pahang_df = df[df['state'] == 'Pahang']
X = pahang_df.drop(['date', 'state', 'cases_new'], axis=1)
y = pahang_df['cases_new']
colnames = X.columns


# Boruta Classifier
rf = RandomForestClassifier(n_jobs=-1, class_weight="balanced_subsample", max_depth=5)
feat_selector = BorutaPy(rf, n_estimators="auto", random_state=1)

# Fit to data
feat_selector.fit(X.values, y.values.ravel())

# Get ranking of features
boruta_score = ranking(list(map(float, feat_selector.ranking_)), colnames, order=-1)
boruta_score = pd.DataFrame(list(boruta_score.items()), columns=['Features', 'Score'])
boruta_score = boruta_score.sort_values("Score", ascending=False)

# Plot with larger font size
sns.set(font_scale=1.5)  # Adjust the font scale factor as needed
sns_boruta_plot = sns.catplot(x="Score", y="Features", data=boruta_score[:], kind="bar",
                              height=14, aspect=1.9, palette='coolwarm')
plt.title("Boruta Feature Selection Pahang", fontsize=24)  
plt.xlabel("Score", fontsize=18)  
plt.ylabel("Features", fontsize=18)  
plt.xticks(fontsize=16)  
plt.yticks(fontsize=16)  
plt.tight_layout()  
plt.show()
```

## Method 2 - Lasso Regularization

```python
relevant_states = ['Pahang', 'Kedah', 'Johor', 'Selangor']
```

```python
# Adapted from https://towardsdatascience.com/feature-selection-with-pandas-e3690ad8504b
    
# Organize plot
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(18, 16))
axes = [ax1, ax2, ax3, ax4]
fig.suptitle("Feature importance (Lasso Model) of various attributes with respect to new cases in the states of "
             "Pahang, Kedah, Johor and Selangor", fontsize=16)

# Perform analysis and plot for each state
for state, ax in zip(relevant_states, axes):
    # Organise data
    relevant_state_data = df[df['state'] == state]
    feature_matrix = relevant_state_data.drop(["date", "state", "cases_new"], axis=1)
    target_data = relevant_state_data["cases_new"]
     
    # Apply regularization: 
    regularization = LassoCV()
    regularization.fit(feature_matrix, target_data)
    
    # Get coefficients from regularization
    # The coefficients tell us how significant an attribute is
    coefficients = pd.Series(regularization.coef_, index = feature_matrix.columns)
    sorted_coefficients = coefficients.sort_values()
    
    # Plot 
    ax.barh(sorted_coefficients.index, sorted_coefficients.values)
    ax.set_title(state)
    ax.set_xlabel("Regularization Coefficient")

plt.show()
```
