```python
#Boxplot
fig, ax = plt.subplots(figsize=(18, 12))
sns.boxplot(data=df, x='state', y='cases_new', ax=ax)
plt.title("New Cases over State")
plt.xlabel("State")
plt.ylabel("New Cases")
```

```python
#Scatterplot

# Create the scatter plot
sns.relplot(x="cases_active", y="cases_unvax", hue="cases_unvax",palette="Oranges", data=df)
plt.xlabel("Active Cases")
plt.ylabel("Unvaccinated Cases")
plt.title("Correlation between Unvaccinated Cases and Active Cases")
plt.show()
```

```python
#Heatmap
correlations = df.corr()
sns.heatmap(correlations)
plt.title("Heat Map")
plt.show()
```

## vi. Finding Outliers within 3-Month Range

```python
# Converting date column to datetime type
df['date'] = pd.to_datetime(df['date'])

#new column representing the 3-month intervals
df['interval'] = pd.PeriodIndex(df['date'], freq='3M')

#unique intervals
intervals = df['interval'].unique()

#remove excluded intervals
filtered_intervals = intervals[::2]

#dataframe filtered based on intervals
filtered_df = df[df['interval'].isin(filtered_intervals)]

#boxplot for the filtered intervals
plt.figure(figsize=(12, 6))
plt.boxplot([filtered_df[filtered_df['interval'] == interval]['cases_new'] for interval in filtered_intervals], 
            positions=range(1, len(filtered_intervals) + 1))

#x-axis labels
labels = [f"{interval.start_time.strftime('%Y-%m')} to {interval.end_time.strftime('%Y-%m')}"
          for interval in filtered_intervals]
plt.xlabel('3-Month Interval')
plt.ylabel('New Cases')
plt.title('Boxplot of New Cases within 3-Month Intervals')
plt.xticks(range(1, len(filtered_intervals) + 1), labels, rotation=45)
plt.show()
```

## vii. Data Visualization using Barchart

```python
#Data Visualization using barchart

# Create lists for the x and y values
Y = list(df.iloc[:, 2])
X = list(df.iloc[:, 1])
plt.figure(figsize=(12, 6))
plt.bar(X, Y, color='#fcb72b')
plt.title("New Cases over States")
plt.xlabel("States")
plt.ylabel("Number of New Cases")

# Rotate x-axis labels
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
