## i. What are the states that exhibit strong correlation with Pahang?

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('cases_state.csv')
df.head()
```

```python
#Filtering the dataset
df_filter = df[df['state'].isin(['Pahang', 'Johor', 'Kedah', 'Kelantan', 'Melaka', 'Negeri Sembilan',
                                   'Pulau Pinang', 'Perak', 'Perlis', 'Selangor', 'Terengganu', 'Sabah',
                                   'Sarawak', 'Wilayah Persekutuan Kuala Lumpur',
                                   'Wilayah Persekutuan Labuan', 'Wilayah Persekutuan Putrajaya'])]
#reshaping using pivot to include states as columns
piv_df = df_filter.pivot(index='date', columns='state', values='cases_new')

#Correlation matrix
corr_matrix = piv_df.corr()
#Correlation coefficients with Pahang
corr_pahang = corr_matrix['Pahang']
#Arrange in descending order
corr_sorted = corr_pahang.sort_values(ascending=False) # ascending false

#A strong positive correlation is bigger than 0.7
strong_positive = corr_sorted[corr_sorted > 0.7]
print("States with strong positive correlation with Pahang:")
print(strong_positive)
#A strong negative correlation is less than -0.7
strong_negative = corr_sorted[corr_sorted < -0.7]
```

## ii. What are the states that exhibit strong correlation with Johor?

```python
df_filter = df[df['state'].isin(['Pahang', 'Johor', 'Kedah', 'Kelantan', 'Melaka', 'Negeri Sembilan',
                                   'Pulau Pinang', 'Perak', 'Perlis', 'Selangor', 'Terengganu', 'Sabah',
                                   'Sarawak', 'Wilayah Persekutuan Kuala Lumpur',
                                   'Wilayah Persekutuan Labuan', 'Wilayah Persekutuan Putrajaya'])]

piv_df = df_filter.pivot(index='date', columns='state', values='cases_new')
corr_matrix = piv_df.corr()
corr_johor = corr_matrix['Johor']
corr_sorted = corr_johor.sort_values(ascending=False)
strong_positive = corr_sorted[corr_sorted > 0.7]
print("States with strong positive correlation with Johor:")
print(strong_positive)
strong_negative = corr_sorted[corr_sorted < -0.7]
```
